```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

URL = 'https://www.independent.co.uk/sport/football/euro-2020-squads-teams-b1848825.html'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
info = []
squad_info = soup.find_all('p')
for squad_ele in squad_info :
  
  players = squad_ele.text.replace('\ufeff','')
  if ':' in players and 'Spain' not in players :
    info.append(players)

  

```

```python
teams = []
team = soup.find_all('div',{'class':"sc-qWSGh iSUKoQ"})
for i in team :
  ele = i.text
  if 'Group' not in ele :
    
   teams.append([ele for i in range(4)])
```

```python
flatten = [item for sublist in teams for item in sublist]
print(len(flatten),len(info))
```

```python
d = {'Country': flatten, 'Meta' : info}
draft = pd.DataFrame(d)
draft['Type'] = draft['Meta'].str.split(':').str[0].str.rstrip('s')
draft['Refined'] = draft['Meta'].str.split(':').str[1].str.lstrip()
draft
```

```python
#draft['Refined'].to_list()
play = []
for i in draft['Refined'].to_list() :
  new = i.split(',')
  for j in new :
    j = j.lstrip()
    play.append(j)
play_dict = {'Players':play}
players_db = pd.DataFrame(play_dict)
players_db['Name']  = players_db['Players'].str.split("(").str[0].str.rstrip('')
players_db['Club'] = players_db['Players'].str.split("(").str[1].str.rstrip(')')
players_db['Club'] = players_db['Club'].str.rstrip(').')
players_db
```

```python
country = []
typea = []
for i in players_db['Players'] :
  for m in range(len(draft)) :

    if i in draft['Refined'].to_list()[m] :
      
      country.append(draft['Country'][m])
      typea.append(draft['Type'][m])
players_db['Country'] = country
players_db['Role'] = typea
players_euro = players_db[['Name','Club','Country','Role']]
players_euro.to_csv('Players Euro.csv',index = False)
players_euro
```
